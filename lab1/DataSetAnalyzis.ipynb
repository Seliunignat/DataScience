```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
dataset_train, dataset_test = train_test_split(data_with_deleted,
test_size=0.2, shuffle=True, stratify=data_with_deleted.DEP_DEL15)

y_train = dataset_train.DEP_DEL15
x_train = dataset_train.drop(columns=['DEP_TIME_BLK','CARRIER_NAME', 'DEPARTING_AIRPORT', 'PREVIOUS_AIRPORT', 'DEP_DEL15'])
y_test = dataset_test.DEP_DEL15
x_test = dataset_test.drop(columns=['DEP_TIME_BLK','CARRIER_NAME', 'DEPARTING_AIRPORT', 'PREVIOUS_AIRPORT', 'DEP_DEL15'])
```

```python
clf = MLPClassifier(random_state=1, max_iter=300, hidden_layer_sizes=(100, 100, 100)).fit(x_train, y_train)
```

```python
result = clf.predict(x_test)
```

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score)
# for i in range (1000, 1100):
#     print(y_test.values[i], end = ' ')
    
# print()
# for i in range (1000, 1100):
#     print(result[i], end = ' ')
# print()

print(accuracy_score(y_test, result))
print(precision_score(y_test, result,  average='macro', zero_division=1))
print(recall_score(y_test, result, average='weighted'))
print(f1_score (y_test, result,  average='weighted'))
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(20, 2, centers=2, random_state=2, 
cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu');
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y);
```
