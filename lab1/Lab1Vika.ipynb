```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
dataset_train, dataset_test = train_test_split(data, test_size=0.2, shuffle=True, stratify = data.DEP_DEL15)
y_train = dataset_train.DEP_DEL15;
y_test = dataset_test.DEP_DEL15;
x_train = dataset_train.drop(columns = ['DEP_TIME_BLK','CARRIER_NAME','DEPARTING_AIRPORT','PREVIOUS_AIRPORT','DEP_DEL15']);
x_test = dataset_test.drop(columns = ['DEP_TIME_BLK','CARRIER_NAME','DEPARTING_AIRPORT','PREVIOUS_AIRPORT','DEP_DEL15']);
```

```python
clf = MLPClassifier(random_state=1, max_iter=300).fit(x_train, y_train)
```

```python
result = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
#accuracy = доля правильных ответов алгоритма

accuracy = accuracy_score(y_test, result)

print (accuracy)
```

```python
from sklearn.metrics import (precision_score, recall_score, f1_score)
#precision = how many selected are relevant, способность отличать этот класс от других классов
#recall = how many relevant are selected,  способность алгоритма обнаруживать данный класс вообще

precision = precision_score (y_test, result, average='weighted', zero_division=1)
recall = recall_score (y_test, result,  average='weighted')
f1 = f1_score (y_test, result,  average='weighted')

print (precision)
print (recall)
print (f1)
```

```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
data = pd.read_csv("full_data_flightdelay.csv")
dataset_train, dataset_test = train_test_split(data, test_size=0.2, shuffle=True, stratify = data.DEP_DEL15)
y_train = dataset_train.DEP_DEL15;
y_test = dataset_test.DEP_DEL15;
x_train = dataset_train.drop(columns = ['DEP_TIME_BLK','CARRIER_NAME','DEPARTING_AIRPORT','PREVIOUS_AIRPORT','DEP_DEL15']);
x_test = dataset_test.drop(columns = ['DEP_TIME_BLK','CARRIER_NAME','DEPARTING_AIRPORT','PREVIOUS_AIRPORT','DEP_DEL15']);

clf = MLPClassifier(random_state=1, max_iter=300).fit(x_train, y_train)

```

```python
result = clf.predict(x_test)
from sklearn.metrics import accuracy_score
from sklearn.metrics import (precision_score, recall_score, f1_score)
accuracy = accuracy_score(y_test, result)
precision = precision_score (y_test, result, average='weighted', zero_division=1)
recall = recall_score (y_test, result,  average='weighted')
f1 = f1_score (y_test, result,  average='weighted')

print (accuracy)
print (precision)
print (recall)
print (f1)
```
